```python
import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1

#pyautogui.click -> clicar
#pyautogui.press -> apertar 1 tecla
#pyautogui.hotkey -> conjunto de telcas
#pyautogui.write -> escreve um texto

#Passo 1: entrar no sistema da empresa (no nosso caso é o link do drive)
pyautogui.hotkey('ctrl', 't')
pyperclip.copy('https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')

while not pyautogui.locateOnScreen('drive.png', confidence=0.9):
    time.sleep(1)

#Passo 2: navegar no sistema e encontrar a base de vendas (entrar na pasta exportar)
pyautogui.click(x=457, y=264, clicks=2)
time.sleep(2)

#Passo 3: fazer o download da base de vendas 
pyautogui.click(x=398, y=341)
pyautogui.click(x=1709, y=153)
pyautogui.click(x=1504, y=563)
time.sleep(5)

```

```python
#Passo 4: importar base de vendas para o python
import pandas as pd
tabela = pd.read_excel(r"C:\Users\luisa\Downloads\Vendas - Dez.xlsx")
display(tabela)
```

```python
#Passo 5: calcular os indicadores da empresa
faturamento = tabela['Valor Final'].sum()
print(faturamento)
quantidade = tabela['Quantidade'].sum()
print(quantidade)
```

```python
#Passo 6: enviar um e-mail para a diretoria com os indicadores 
#abrir aba
pyautogui.hotkey('ctrl', 't')

#entrar no link do e-mail - https://mail.google.com/mail/u/0/#inbox
pyperclip.copy('https://mail.google.com/mail/u/0/#inbox')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')
time.sleep(5)

#clicar no botão escrever
pyautogui.click(x=93, y=172)

#preencher as informações do e-mail
pyautogui.write('pythonimpressionador@gmail.com')
pyautogui.press('tab') #selecionar o e-mail

pyautogui.press('tab')
pyperclip.copy('Relatório de vendas') #pular para o campo de assunto
pyautogui.hotkey('ctrl','v')

pyautogui.press('tab') #pular para o campo de corpo do e-mail

texto = f'''
Prezados, 

segue o relatório de vendas.
Faturamento: R${faturamento:,.2f}
Quantidade de produtos vendidos: {quantidade}


Qualquer dúvida estou à disposição.

Att.,
Luisa

'''


#formatação de números
pyperclip.copy(texto)
pyautogui.hotkey('ctrl', 'v')

#anexar o arquivo
arquivo = r'C:\Users\luisa\Downloads\Vendas - Dez.xlsx'
pyautogui.click(x=1421, y=1004)
time.sleep(2)
pyperclip.copy(arquivo)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')


#enviar o e-mail
pyautogui.hotkey('ctrl', 'enter')
```
